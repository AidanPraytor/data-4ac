We can use spatial joins as a way to summarize the number of sites per polygon (counties or watersheds)

```python
#https://towardsdatascience.com/how-to-easily-join-data-by-location-in-python-spatial-join-197490ff3544
#http://andrewgaidus.com/leaflet_webmaps_python/ example of join and use with folium
#Spatial join Points to polygons [https://geopandas.org/mergingdata.html?highlight=count%20points]
#This example gets counts of points in each county using a two part process,
#first join counties info to sites then 
cntyjoin_sites = gpd.sjoin(sites, counties, op = 'within') 

cntyjoin_sites.head()
```

```python
#group pandas style to count points in counties
grouped = cntyjoin_sites.groupby('NAMELSAD').size()
df = grouped.to_frame().reset_index()
df.columns = ["NAMELSAD", "sites_count"]

#merge the count pandas dataframe to counties polygon
merged_areas = counties.merge(df, on="NAMELSAD", how= 'outer')

#plot merged areas based on quantiles of sites_count
#this could be a homework to reuse mapclassify

#attempt at alternative https://gis.stackexchange.com/questions/306674/geopandas-spatial-join-and-count
#dfpivot = pd.pivot_table(countyjoin,index='NAMELSAD')
#dfpivot.columns = dfpivot.columns.droplevel()
#dfpolynew = polys.merge(dfpivot, how='left',on='NAMELSAD')
```
